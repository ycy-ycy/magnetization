```python
mList = [np.load('%s.npy'%(str(i).zfill(2))) for i in range(41)]
mTarget = np.load('mTarget.npy')
weight = np.load('weight.npy')
points = np.column_stack(np.where(weight == 1))
listX, listY = [],[]
mPixelX, mPixelY = weight.shape
NofP = len(points)
for i in range(NofP):
    listX.append(points[i][0])
    listY.append(points[i][1])
listX, listY = np.array(listX), np.array(listY)

def cosMap(m,mT):
    result = []
    for i in range(NofP):
        mv = m[listX[i],listY[i],:]
        tv = mT[listX[i],listY[i],:]
        m2 = np.sum(mv**2)
        t2 = np.sum(tv**2)
        if (m2>0) and (t2>0):
            cosi = np.sum(mv*tv)/np.sqrt(m2*t2)
        else:
            cosi = 0
        result.append(cosi)
    
    return np.mean(result)
```

```python
lines = data.split('\n')
groups = [lines[i:i+3] for i in range(0, len(lines), 3)]
```

```python
parsed_data = []

for group in groups:
    round_info = group[0].split(',')
    distance = float(round_info[1].split('=')[1])
    xShift = float(round_info[2].split('=')[1])
    yShift = float(round_info[3].split('=')[1])
    loss = float(round_info[4].split('=')[1])
    
    values = list(map(float, group[1].split()))
    
    round_time = float(group[2].split(':')[1].strip())
    
    parsed_data.append({
        'distance': distance,
        'xShift': xShift,
        'yShift': yShift,
        'loss': loss,
        'values': values,
        'round_time': round_time
    })
```

```python
rounds = list(range(41))
distances = [i['distance'] for i in parsed_data]
losses = [i['loss'] for i in parsed_data]
assimilarities = [i['values'][0] for i in parsed_data]
cosis = [cosMap(i,mTarget) for i in mList]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

axs[0][0].scatter(rounds, distances, color='blue', marker='x', label='distance')
axs[0][0].axhline(y=12, color='red', linestyle='--', label='true distance')
axs[0][0].set_title('Rounds vs Distance')
axs[0][0].set_xlabel('Rounds')
axs[0][0].set_ylabel('Distance')
axs[0][0].legend()

axs[0][1].scatter(rounds, losses, color='green', marker='x', label='loss')
axs[0][1].set_title('Rounds vs Loss')
axs[0][1].set_xlabel('Rounds')
axs[0][1].set_ylabel('Loss (log scale)')
axs[0][1].set_yscale('log')
axs[0][1].legend()

axs[1][0].scatter(rounds, assimilarities, color='purple', marker='x', label='assimilarity')
axs[1][0].axhline(y=1, color='orange', linestyle='--', label='assimilarity=1')
axs[1][0].set_title('Rounds vs Assimilarity')
axs[1][0].set_xlabel('Rounds')
axs[1][0].set_ylabel('Values (log scale)')
axs[1][0].set_yscale('log')
axs[1][0].legend()

axs[1][1].scatter(rounds, cosis, color='orange', marker='x', label=r'cos$\theta$')
axs[1][1].axhline(y=1, color='cyan', linestyle='--', label=r'cos$\theta$=1')
axs[1][1].axhline(y=0, color='blue', linestyle='--', label=r'cos$\theta$=0')
axs[1][1].set_title(r'Rounds vs cos$\theta$')
axs[1][1].set_xlabel('Rounds')
axs[1][1].set_ylabel(r'cos$\theta$')
axs[1][1].legend()

plt.tight_layout()

plt.savefig('result.png')

plt.show()
```
